```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
import pandas as pd
import numpy as np
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,TimeoutException, ElementClickInterceptedException
import undetected_chromedriver as uc
import json
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
link_produk = 'https://sonushub.id/category/inverter-kb7cc?page=1'
path = r"C:/Users/KBIBKU- BNBYTO/Documents/Scraping Data Komponen plts/chromedriver-win64/chromedriver.exe"
```

```python
options = uc.ChromeOptions()
options.add_argument("--start-maximized")

driver = uc.Chrome(options=options)

driver.set_window_size(1300,800)
driver.get(link_produk)
time.sleep(5)
```

```python
halaman = 1
product_data = []
while True:
    print(f"Scraping halaman ke-{halaman}")
    produk_list = driver.find_elements(By.XPATH, "//div[contains(@class, 'v-col v-col-12 minw-0 flex-shrink-0')]")
    print(f"Jumlah elemen ditemukan: {len(produk_list)}")
    for i, produk in enumerate(produk_list, start=1):
        print(f"Proses data ke-{i}")
        try:
            # Tangkap teks dan atribut dengan find_elements
            nama = produk.find_elements(By.XPATH, ".//h5[contains(@class, 'opacity-60 fw-400 mb-2 lh-1-6 fs-13')]")
            harga = produk.find_elements(By.XPATH, ".//div[contains(@class, 'fs-16 mb-2')]")
            link = produk.find_elements(By.XPATH, ".//a")
            # Ambil data jika ada, jika kosong beri nilai None
            nama_text = nama[0].text if nama else "None"
            harga_text = harga[0].text if harga else "None"
            link_href = link[0].get_attribute('href') if link else "None"
            # Cetak hasil
            print(f"Nama: {nama_text}")
#             print(f"Link: {link_href}")
#             print('-------')
            product_data.append({
            "nama produk": nama_text,
            "harga": harga_text,
            "link": link_href
            })  
        except Exception as e:
            print(f"Error memproses data ke-{i}: {e}")
            
    try:
        tombol_next = driver.find_element(By.XPATH, "//li[@class='v-pagination__next']/button")

        # Periksa apakah tombol disabled
        aria_disabled = tombol_next.get_attribute('aria-disabled')

        if aria_disabled == 'true':
            print("Tombol halaman selanjutnya tidak aktif (disabled). Scraping selesai.")
            break

        # Klik jika tidak disabled
        tombol_next.click()
        print(f"Berhasil klik tombol halaman selanjutnya.")
        halaman += 1
        time.sleep(3)
    except NoSuchElementException:
        print("Tombol halaman selanjutnya tidak ditemukan.")
        # Coba cari dengan class name yang lengkap dari HTML
        try:
            tombol_next = driver.find_element(By.CSS_SELECTOR, "li.v-pagination_next button")
            # Alternatif lain
            # tombol_next = driver.find_element(By.CSS_SELECTOR, "button[aria-label='$vuetify.pagination.ariaLabel.next']")
            tombol_next.click()
            print(f"Berhasil klik tombol halaman selanjutnya dengan selector alternatif.")
            halaman += 1
            time.sleep(3)
        except Exception as e:
            print(f"Gagal dengan selector alternatif juga: {e}")
            break
    except Exception as e:
        print(f"Error saat menangani pagination: {e}")
        break
```

## Data deskripsi produk

```python
deskripsi_data = [] 
for idx, url in enumerate(df['link']):
    print(f"\nMemulai scraping untuk produk {idx + 1}: {url}")
    driver.get(url)
    # Tunggu halaman dimuat sepenuhnya
    time.sleep(5)
    
    try:
        # Gunakan WebDriverWait untuk memastikan elemen sudah muncul
        from selenium.webdriver.support.ui import WebDriverWait
        from selenium.webdriver.support import expected_conditions as EC
        
        # Tunggu sampai elemen deskripsi muncul di DOM
        deskripsi = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'v-expansion-panel-text__wrapper')]"))
        )
        
        # Ambil teks dari elemen yang ditemukan
        deskripsi_text = deskripsi.text
        
        # Cetak hasil
#         print(f"deskripsi: {deskripsi_text}")
#         print('-------')
        deskripsi_data.append({
            "no": idx+1,
            "spesifikasi": deskripsi_text
        })
    except Exception as e:
        print(f"Error saat scraping produk {idx + 1}: {e}")
        deskripsi_data.append({"spesifikasi": "None"})
```

```python
import re

pola_hapus = [
    r'\nMinimal order.*',
    r'\nKonfirmasi sebelum order',
    r'\nHubungi admin untuk.*'
]

# fungsi general untuk membersihkan spesifikasi
def bersihkan_spesifikasi(teks, daftar_pola):
    for pola in daftar_pola:
        teks = re.sub(pola, '', teks)
    return teks.strip()
```

```python
spesifikasi['spesifikasi_formatted'] = spesifikasi['spesifikasi'].apply(lambda x: bersihkan_spesifikasi(x, pola_hapus))
```
